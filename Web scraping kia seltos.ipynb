```python
driver = webdriver.Chrome('C:/Users/HP/Videos/chromedriver_win32/chromedriver.exe')
driver.get("https://www.cardekho.com/kia/seltos/user-reviews")
```

```python
def count_star(ratingobj):
    i=0
    for spans in ratingobj.find('span'):
        print(spans)
        if 'star-full-fill' in str(spans):
            i=i+1
    return(i)
```

```python
Text = []
Date = []
#Rating =[]
Name = []
star_rating =[]

for i in range(1,100):
    _url = "https://www.cardekho.com/kia/seltos/user-reviews/{}?subtab=latest".format(i)
    _url = "https://www.cardekho.com/kia/seltos/user-reviews"
    driver.get(_url)
    content = driver.page_source
    soup = BeautifulSoup(content)
    for a in soup.findAll('div',{'class':'readReviewBox'}):
        textObj=a.find('div', attrs={'class':'contentspace'})
        dateObj=a.find('div', attrs={'class':'date'})
        ratingobj=a.find('div', attrs={'class':'starRating'})
        star = count_star(ratingobj)
        nameObj=a.find('div', attrs={'class':'name'})
        
        if textObj and dateObj and ratingobj and nameObj:
            Text.append(textObj.text)
            Date.append(dateObj.text)
            star_rating.append(star)
            Name.append(nameObj.text)

        df = pd.DataFrame({'Review Text':Text,'Review Date':Date,'Review Rating':star_rating,'Reviewer Name':Name}) 
        df
```
